```python
sns.countplot(data=df,x="Sex")
plt.title("Distribution by sex")
plt.show()
```

```python
sns.histplot(data=df,x="Age",bins=10,kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(data=df,x="Height",bins=10,kde=True)
plt.title("Height Distribution")
plt.show()
```

```python
sns.histplot(data=df, x="Weight",bins=20, kde=True)
plt.title("Weight Distribution")
plt.show()
```

```python
sns.countplot(data=df, x="Medal")
plt.title("Medal Distribution")
plt.show()
```

```python
sns.countplot(data=df,x="Year",hue="Medal")
plt.title("Medal Distribution over the years")
plt.xticks(rotation=45) 
plt.show()
```

```python
year_avg_age=df.groupby("Year")["Age"].mean()
print(year_avg_age)
```

```python
sport_median_height=df.groupby("Sport")["Height"].median()
print(sport_median_height.max())
```

```python
sport_median_height[sport_median_height==190]
```

```python
country_gender_count=df.groupby(["NOC","Sex"])["ID"].count()
print(country_gender_count)
```

```python
country_gold_medal=df[df["Medal"]=="Gold"].groupby("NOC")["Medal"].count()
print(country_gold_medal)
```

```python
print(country_gold_medal.max())
```

```python
country_gold_medal[country_gold_medal==747]
```

```python
sport_gender_avg_wt=df.groupby(["Sport","Sex"])["Weight"].mean()
print(sport_gender_avg_wt)
```

```python
print(sport_gender_avg_wt["Wrestling"]["F"])
```

```python
sport_event_count=df.groupby("Sport")["Event"].nunique().sort_values(ascending=False)
sport_event_count.plot(kind="bar")
plt.title("Number of unique event per sport")
plt.xlabel("Sport")
plt.ylabel("Number of unique events")
plt.xticks(rotation=45)
plt.show()
```

```python
year_participant_count=df.groupby("Year")["ID"].nunique()
year_participant_count.plot(kind="bar")
plt.title("Number of Participants Over the Years")
plt.xlabel("Years")
plt.ylabel("Number of Participant")
plt.show()
```

```python
country_avg_age=df.groupby("NOC")["Age"].mean().sort_values(ascending=False)
country_avg_age.tail(10).plot(kind="bar")
plt.title("Countries with Lowest Average age of Participants")
plt.ylabel("Average Age")
plt.xlabel("Countries")
plt.show()
```

```python
sns.boxplot(data=df,x='Season',y="Age")
plt.title("Distribtuion of Age by Season")
plt.xlabel("Season")
plt.ylabel("Age")
plt.show()
```

```python
sns.violinplot(data=df,x="Medal",y="Height",palette="Set2",hue="Medal")
plt.xlabel("Medal")
plt.ylabel("Height")
plt.title("Medal Vs Height of Athelete Plot")
plt.show()
```

```python
most_medal_country=df["NOC"].value_counts().idxmax()
print(most_medal_country)
```

```python
tallest_athelete=df[df["Height"]==df["Height"].max()]
print(tallest_athelete)
```

```python
heaviest_athelete=df[df["Weight"]==df["Weight"].max()]
print(heaviest_athelete[["Name","ID","Sport","Weight"]])
      
```

```python
lightest_athelete=df[df["Weight"]==df["Weight"].min()]
print(lightest_athelete[["Name","ID","Sport","Weight"]])
```

```python
sns.scatterplot(data=df,x="Height",y="Weight",hue="Medal",palette="Set2")
plt.title("Distribution of Medal vs Height Vs Weight in Scatter plot")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
          
          
```

```python
medal_by_country_year=df.pivot_table(index="NOC",columns="Year",values="Medal",aggfunc="count")
sns.heatmap(medal_by_country_year,cmap="YlGnBu",linewidth=0.5)
plt.xlabel("Year")
plt.ylabel("Country")
plt.xticks(rotation=45)
plt.show()
```
